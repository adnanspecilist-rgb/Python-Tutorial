## **Literal:**
Literal is a raw data given  a variale in python
1. Numeric Literal
2. String literal
3. oolean Literal
4. Special Literal


```python
#Numeric Literal
a= 0b1010 #binary Literal ->output decimal
b=100 #decimal Literal
c=0o310# Octal Literal where 310 octal numer
d=0x12c #Hexadecimal Literal 12c is hexadecimal numbers

#Flat Literal
flat1=10.5
flat2= 1e-3

#Complex Literal
x= 3+5j

print(a,b,c)
print(flat1,flat2)
print(x,x.imag,x.real)
```

```python
#String Literal
string = "Adnan"
multi_string=  " " "use for long paragraph" " "
unicode = "\U0001f600\U0001F606\U0001F923" #Emoji
print(multi_string)
print(unicode)

#Boolean Literal
a=True+4 # 1+4
b=False+10 # 0+10
print(a,b)

#Special Literal
a= None # use for declaration
print(a)
```
