```python
# Practice Question
cemail="adnan@gmail.com"
cpassword="12345"
email=input("Enter Email")

if '@' in email:
    password = input("Enter password") # Removed int() to keep password as string for comparison

    if email == cemail and password == cpassword:
        print("Welcome")
    elif email == cemail and password != cpassword:
        print("Incorrect Password")
        password = input("Enter password again") # Reassign password variable (now a string)
        if password == cpassword: # Nested if statement
            print("Finally correct")
        else:
            print("Still incorrect")
    else: # This 'else' covers cases where email is not cemail, but format was fine.
        print("Invalid")
else: # This 'else' covers cases where '@' is not in email (invalid email format).
    print("Invalid")
```
